```python
driver = webdriver.Chrome('C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe')
url = 'http://topis.seoul.go.kr/refRoom/openRefRoom_1_3.do'
driver.get(url)
```

```python
# 2017년 선택
year2017 = driver.find_element_by_xpath('//*[@id="selYear"]/option[7]')
year2017.click()
```

```python
# 지역 선택
# 종로구 ~ 강동구 option[i] 돌리기 4 ~ 22
# 종로구 선택
selReg = driver.find_element_by_xpath('//*[@id="selRegCd"]/option[4]')
selReg.click()
```

```python
# 월 선택
# 1 ~ 12 월까지 option[i] 돌리기 1 ~ 12
# for i in range(1,13) :
# selMon = driver.find_element_by_xpath('//*[@id="selMon"]/option['+str(i)+']')
selMon = driver.find_element_by_xpath('//*[@id="selMon"]/option[1]')
selMon.click()
```

```python
selReg = driver.find_element_by_xpath('//*[@id="selRegCd"]/option[5]')
selReg.click()
```

```python
# 오전, 낮, 오후, 전일, 전체, 시간대별 option[i] 돌리기 1 ~ 22
# 이 과정에서 오류가 한번이라도 나면 지역이 전체로 바뀜!!!
# 그래서 for문에 지역도 넣었음
for i in range(1,23) :
    # 종로구 선택
    selReg = driver.find_element_by_xpath('//*[@id="selRegCd"]/option[4]')
    selReg.click()
    # 시간 선택
    selTime = driver.find_element_by_xpath('//*[@id="selTimeGrp"]/option['+str(i)+']')
    selTime.click()
    sleep(5)
    # 검색
    search = driver.find_element_by_xpath('//*[@id="content"]/table/tbody/tr/td/table/tbody/tr/td[5]/button')
    search.click()
    sleep(5)
    # 엑셀 다운
    excel = driver.find_element_by_xpath('//*[@id="content"]/table/tbody/tr/td/table/tbody/tr/td[6]/button[1]')
    excel.click()
    sleep(5)
    
# for문이 5번 이상 돌면 404 error가 떠서 실패!
```

```python
traffic = pd.read_excel('C:\\Users\jeens\OneDrive\Desktop\통행속도\차량통행속도.xls', encoding ='ANSI')
traffic = traffic.drop('날씨', axis = 1)
traffic = traffic.drop('최고온도(℃)', axis = 1)
traffic = traffic.drop('최저온도(℃)', axis = 1)
traffic.columns = ['일자', '0']
traffic.head()
```

```python
for i in range(1,22) :
    traffic2 = pd.read_excel('C:\\Users\jeens\OneDrive\Desktop\통행속도\차량통행속도 ('+str(i)+').xls', encoding ='ANSI')
    traffic2 = traffic2.drop('날씨', axis = 1)
    traffic2 = traffic2.drop('최고온도(℃)', axis = 1)
    traffic2 = traffic2.drop('최저온도(℃)', axis = 1)
    traffic2.columns = ['일자', str(i)]
    traffic = pd.merge(traffic, traffic2, on='일자')
```
